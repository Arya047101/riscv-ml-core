```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Load and Preprocess Data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize and Reshape for CNN (requires 4D input: batch, height, width, channels)
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# 2. Build the CNN Model
model_cnn = models.Sequential([
    # Convolutional layer: extracts features into a 26x26 output
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    # Max Pooling layer: reduces 26x26 to 13x13 (4x data reduction)
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 3. Compile and Train
model_cnn.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("Training CNN with Max Pooling...")
model_cnn.fit(x_train, y_train, epochs=5, batch_size=32)

# 4. Evaluate
test_loss, test_acc = model_cnn.evaluate(x_test, y_test, verbose=0)
print(f'\nCNN Test Accuracy: {test_acc:.4f}')
model_cnn.summary()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Load and Preprocess Data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values (0-255 to 0-1)
x_train, x_test = x_train / 255.0, x_test / 255.0

# 2. Build the MLP Model
# Matches the uploaded notebook structure
model_mlp = models.Sequential([
    layers.Flatten(input_shape=(28, 28)), # Converts 28x28 to 784x1
    layers.Dense(128, activation='relu'), # Hidden layer with 128 neurons
    layers.Dense(10, activation='softmax') # Output layer for 10 digits
])

# 3. Compile and Train
model_mlp.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("Training Simple MLP...")
model_mlp.fit(x_train, y_train, epochs=5, batch_size=32)

# 4. Evaluate
test_loss, test_acc = model_mlp.evaluate(x_test, y_test, verbose=0)
print(f'\nMLP Test Accuracy: {test_acc:.4f}')
model_mlp.summary()
```
